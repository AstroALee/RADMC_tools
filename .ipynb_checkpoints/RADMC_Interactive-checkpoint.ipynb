```python
# Import relevant modules 
import RADMC_DataAnalysis as DA
import numpy as np
import matplotlib.pyplot as plt

from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
# Load the relevant data set
#
# header = dictionary of info 
# df = Pandas dataframe of intensities, e.g., df[3] gives pixel 3's data at all wavelengths
header,df = DA.LoadData('image.out')
```

```python
# pixels to plot (make a list, even if single value)
pixs = [1000,1500,1800,2000]

# normalize? 
yes_norm = 1

norm_val = 1.0
if(yes_norm):
    norm_val = max(df.max())
    print(norm_val)

for pix in pixs:
        (df[pix]/norm_val).plot() 
        
DA.plt.legend();
DA.plt.ylim([0,1e-14]);
if(yes_norm): DA.plt.ylim([0,1]);
```

```python
# Moment map

# moment to plot
moment = 0 

# normalize?
yes_norm = 1

Nx = header["Nx"]
Ny = header["Ny"]
img = np.array(DA.Find_IntMomentMap(df,moment)).reshape(Nx,Ny)

norm_value = 1.0
if(yes_norm):
    norm_value = img.max()
    
DA.plt.imshow(img/norm_value,origin='lower')
DA.plt.colorbar();

mstr = 'Intensity   ' + '$\int I_{v} v^' + str(moment) + ' dv$';
if(yes_norm):
    mstr = 'Normed ' + mstr

DA.plt.title(mstr);
```
